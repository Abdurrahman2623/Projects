```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#df.remove duplicates
df.drop_duplicates()
df
```

```python
#replace values in columns
df["Near_Location"].replace(1,'Yes', inplace = True)
df["Near_Location"].replace(0,'No', inplace = True)
df["Partner"].replace(1,'Yes', inplace = True)
df["Partner"].replace(0,'No', inplace = True)
df["Promo_friends"].replace(1,'Yes', inplace = True)
df["Promo_friends"].replace(0,'No', inplace = True)
df["Phone"].replace(1,'Yes', inplace = True)
df["Phone"].replace(0,'No', inplace = True)
df["Group_visits"].replace(1,'Yes', inplace = True)
df["Group_visits"].replace(0,'No', inplace = True)
df["Churn"].replace(1,'Yes', inplace = True)
df["Churn"].replace(0,'No', inplace = True)
df["gender"].replace(1,'Male', inplace = True)
df["gender"].replace(0,'Female', inplace = True)
df
```

```python
x = pd.get_dummies(df.drop(['Churn'], axis = 1))
y = df['Churn'].apply(lambda x:1 if x== 'Yes' else 0)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = .2, random_state=42)
```
